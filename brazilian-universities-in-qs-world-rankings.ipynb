# Improvement of Brazilian Universities

Analyzing the difference of the Brazilian Universities in the QS World University Rankings between the years of 2023 and 2024 

## Setting up the environment

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Getting the Data

```python
ranking_path = './dataset/2024 QS World University Rankings 1.1 (For qs.com).csv'

ranking_data = pd.read_csv(ranking_path).drop(axis=0, index=0)
```

```python
ranking_data.head()
```

## How is the ranking distribuited around the world?

```python
plt.figure(figsize=(10,20))
sns.set_style('darkgrid')
sns.histplot(data=ranking_data, y='Country Code')
```

It's possible to say that the top 3 countries that have more universities in the ranking are:
1. US
2. UK
3. CN

## Analyzing Brazilian Universities
We can see that Brazil stays between 25 and 50 universities, let's get the exact number.

```python
br_names_ranking_data = ranking_data[(ranking_data['Country Code'] == 'BR')]
br_names_ranking_data.head()
```

```python
br_names_ranking_data['Institution Name'].count()
```

Brazil has 35 universities in the ranking.
